```python
import numpy as np
from scipy import linalg
```

```python
def f(x):
    return (np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2))
```

```python
a = np.array([[1,1],[1,15]])
b = np.array([f(1),f(15)])
w = linalg.solve(a,b)
w
```

```python
%matplotlib inline
from matplotlib import pylab as plt
x = np.arange(1,30,0.3)
vfunc = np.vectorize(f) 
y = vfunc(x)
plt.plot(x,y)

ay= w[0] + w[1]*x
plt.plot(x,ay)
plt.show()
```

```python
a = np.array([[1,1,1],[1,8,64],[1,15,225]])
b = np.array([f(1),f(8),f(15)])
w = linalg.solve(a,b)

x = np.arange(1,30,0.3)
vfunc = np.vectorize(f) 
y = vfunc(x)
plt.plot(x,y)

ay= w[0] + w[1]*x + w[2]*x**2
plt.plot(x,ay)
plt.show()
```

```python
a = np.array([[1,1,1,1],[1,4,16,64],[1,10,100,1000],[1,15,225,3375]])
b = np.array([f(1),f(4),f(10),f(15)])
w = linalg.solve(a,b)

x = np.arange(1,30,0.3)
vfunc = np.vectorize(f) 
y = vfunc(x)
plt.plot(x,y)

ay= w[0] + w[1]*x + w[2]*x**2 + w[3]*x**3
plt.plot(x,ay)
plt.show()
```

```python
file_sol = open('submission-2.txt','w')
w =np.around(w,decimals=3)
for num in w:
    file_sol.write(str(num)+' ')
    print(type(num))
file_sol.close()
w
```
